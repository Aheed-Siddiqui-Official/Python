```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns   
```

```python
data = pd.read_csv(r"./play_golf_data.csv")
```

```python
data.describe()
```

```python
df = data.drop(columns=['Unnamed: 0'])
df
```

```python
# Data Cleaning
df_clean = df[df.isnull().sum(axis=1)<4].reset_index(drop=True).copy()
```

```python
df_clean.head()
```

```python
# This code fills in missing values (NaNs) in two specific columns so you don't have "holes" in your data.

HmMean = df_clean['Humidity'].mean()
HdMode = df_clean['Holiday'].mode()[0]
df_clean['Humidity'] = df_clean['Humidity'].fillna(HmMean)
df_clean['Holiday'] = df_clean['Holiday'].fillna(HdMode)
df_clean.head()
```

```python
# Linear Interpolation
df_clean['Temp'] = df_clean['Temp'].interpolate(method='linear')
df_clean.head()
```

```python
# Forward/Backward Clean
df_clean['Outlook'] = df_clean['Outlook'].ffill()
df_clean.head()
```

```python
# .fillna(-1): It looks for every missing value (NaN) in the 'Wind' column and replaces it with the number -1.
df_clean['Wind'] = df_clean['Wind'].fillna(-1)
```
