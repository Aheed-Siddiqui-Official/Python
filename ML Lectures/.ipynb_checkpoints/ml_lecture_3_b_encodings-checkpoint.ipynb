```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"./emp.csv")
```

```python
df.head()
```

```python
# Target Encoding

df2 = df.copy()
```

```python
sorted(df2["job"].unique())
```

```python
df2['job'].value_counts()
```

```python
cols_nums = {"job": {'ANALYST':3, 'CLERK':5, 'MANAGER':2, 'PRESIDENT':1, 'SALESMAN':4}}
```

```python
df2 = df2.replace(cols_nums)
df2.head()
```

```python
def salCatg(salary):
    if salary>=3000:
        r=1
    elif salary>=1500:
        r=2
    else:
        r=3
    return r
```

```python
df2['sal'] = df2['sal'].apply(salCatg)
df2.head()
```

```python
# Label Encoding applies to category data types

df3 = df.copy()
df3.dtypes
```

```python
df3['job'] = df3["job"].astype('category')
df3.dtypes
```

```python
sorted(df3["job"].unique())
```

```python
df3
```

```python
df3['job1'] = df3['job'].astype('category').cat.codes
df3.head()
```

```python
df3['comm1'] = df3['comm1'].astype('category').cat.codes
df3.head()
```
