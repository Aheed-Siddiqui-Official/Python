```python
import pandas as pd
import numpy as np

df = pd.read_csv("../DataSets/PlayGolfData.csv")
```

```python
df_original = df.copy()
```

# LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder

df_label = df_original.copy()

le = LabelEncoder()
df_label['Outlook'] = le.fit_transform(df_label['Outlook'])

df_label.head()
print(df_label.columns)
```

# ONE-HOT ENCODING


```python

df_onehot = df_original.copy()

df_onehot = pd.get_dummies(df_onehot, columns=['Outlook'])

df_onehot.head()
print(df_onehot.columns)
```

# BINARY ENCODING

```python


import category_encoders as ce

df_binary = df_original.copy()

encoder = ce.BinaryEncoder(cols=['Outlook'])
df_binary = encoder.fit_transform(df_binary)

df_binary.head()
print(df_binary.columns)
```

# TARGET ENCODING

```python


df_target = df_original.copy()

encoder = ce.TargetEncoder(cols=['Outlook'])
df_target['Outlook'] = encoder.fit_transform(df_target['Outlook'], df_target['Crowdedness'])

df_target.head()
```

# ORDINAL ENCODING

```python


from sklearn.preprocessing import OrdinalEncoder

df_ordinal = df_original.copy()

oe = OrdinalEncoder(categories=[['Low','High','Extreme']])
df_ordinal[['Temperature']] = oe.fit_transform(df_ordinal[['Temperature']])

df_ordinal.head()
```

# CYCLIC ENCODING

```python


df_cyclic = df_original.copy()

month_map = {'Jan':1,'Feb':2,'Mar':3,'Apr':4}
df_cyclic['Month_num'] = df_cyclic['Month'].map(month_map)

df_cyclic['Month_sin'] = np.sin(2*np.pi*df_cyclic['Month_num']/12)
df_cyclic['Month_cos'] = np.cos(2*np.pi*df_cyclic['Month_num']/12)

day_map = {'Mon':1,'Tue':2,'Wed':3,'Thu':4,'Fri':5,'Sat':6,'Sun':7}
df_cyclic['Day_num'] = df_cyclic['Weekday'].map(day_map)

df_cyclic['Day_sin'] = np.sin(2*np.pi*df_cyclic['Day_num']/7)
df_cyclic['Day_cos'] = np.cos(2*np.pi*df_cyclic['Day_num']/7)

df_cyclic.head()
```
