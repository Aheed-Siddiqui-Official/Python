# Question 13

```python
# Dataset - House Price Prediction
X = np.array([650, 785, 1200, 1375, 1500, 1650, 1800, 1950,
              2100, 2250, 2400, 2550, 2700, 2850, 3000, 3200,
              3400, 3600, 3850, 4700]).reshape(-1, 1)   # Feature: House size

y = np.array([3200, 3650, 5700, 5900, 6050, 6700, 7150, 7800,
              8250, 8900, 9400, 9950, 10400, 10900, 11500, 12300,
              13050, 13700, 14600, 15400]).reshape(-1, 1)  # Target: Price in thousands PKR
```

```python
# Feature Scaling (Very Important to avoid overflow)
X_mean = X.mean()
X_std = X.std()
X_scaled = (X - X_mean) / X_std
```

```python
# Add bias term (column of ones)
X_b = np.c_[np.ones((len(X), 1)), X_scaled]
```

```python
# Hyperparameters
epochs = 1000
learning_rate = 0.1        # better learning rate after scaling
```

```python
# Initialize parameters
theta = np.zeros((2, 1))   # theta[0] = bias, theta[1] = slope
```

```python
for epoch in range(epochs):
    # Predictions
    y_pred = X_b.dot(theta)
    
    # Error
    error = y_pred - y
    
    # Gradient
    gradients = (2 / len(X)) * X_b.T.dot(error)
    
    # Update theta
    theta = theta - learning_rate * gradients
    
    if epoch % 200 == 0:
        mse = np.mean(error ** 2)
        print(f"Epoch {epoch:4d} | MSE = {mse:.2f}")

print("\nTraining Completed Successfully!")
print(f"Final Intercept (b) : {theta[0][0]:.4f}")
print(f"Final Slope (w)     : {theta[1][0]:.4f}")
```

```python
# Make Prediction for a new house (2500 sq ft)
new_size = 2500
new_size_scaled = (new_size - X_mean) / X_std
new_X = np.array([[1, new_size_scaled]])          # add bias
predicted_price = new_X.dot(theta)[0][0]

print(f"\nPredicted price for {new_size} sq ft house: {predicted_price:.2f} thousand PKR")
```

# Summary

Implemented Linear Regression manually using Batch Gradient Descent from scratch (without using scikit-learn).

Used a house price prediction dataset where the feature is house size (in square feet) and the target is price (in thousands of PKR).
    
Applied feature scaling to prevent overflow errors and improve training stability.
    
Trained the model over 1000 epochs and observed how Mean Squared Error (MSE) decreased during training.
    
Successfully learned the intercept (bias) and slope (weight).

Made a prediction for a new house of 2500 sq ft.


# Question 14

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
import pandas as pd
```

```python
# Dataset - Student Health & Fitness Data
# Features: Study Hours per day, Sleep Hours, Exercise Minutes per day
X = np.array([
    [4.5, 6.0, 30],
    [6.0, 7.5, 45],
    [3.0, 5.5, 20],
    [7.5, 8.0, 60],
    [5.5, 6.5, 35],
    [2.5, 5.0, 15],
    [8.0, 7.0, 50],
    [4.0, 8.5, 40],
    [6.5, 6.0, 55],
    [5.0, 7.0, 25]
])

# Targets: 
# Target 1: Attention Score (out of 100)
# Target 2: Stress Level (scale 1-100)
# Target 3: Memory Retention Score (out of 100)
y = np.array([
    [65, 75, 70],
    [85, 45, 88],
    [55, 85, 60],
    [92, 35, 95],
    [78, 60, 80],
    [50, 90, 55],
    [88, 40, 90],
    [72, 65, 75],
    [82, 50, 85],
    [68, 70, 72]
])
```

```python
# Train Multi-output Linear Regression
model = LinearRegression()
model.fit(X, y)
```

```python
# Predictions
y_pred = model.predict(X)
```

```python
# Evaluation
print("R² Scores for each target:")
target_names = ['Attention Score', 'Stress Level', 'Memory Retention']
for i, target_name in enumerate(target_names):
    r2 = r2_score(y[:, i], y_pred[:, i])
    mae = mean_absolute_error(y[:, i], y_pred[:, i])
    print(f"{target_name:20}: R² = {r2:.4f} | MAE = {mae:.2f}")
```

```python
# Show Coefficients
print("\nModel Coefficients (for each target):")
coef_df = pd.DataFrame(model.coef_,
                       columns=['Study Hours', 'Sleep Hours', 'Exercise Minutes'],
                       index=target_names)
print(coef_df.round(4))
```

```python
# Example Prediction
new_student = np.array([[7.0, 8.0, 50]])   # New student data
prediction = model.predict(new_student)
print("\nPrediction for new student (Study=7hrs, Sleep=8hrs, Exercise=50min):")
print(f"Attention Score     : {prediction[0][0]:.1f}")
print(f"Stress Level        : {prediction[0][1]:.1f}")
print(f"Memory Retention    : {prediction[0][2]:.1f}")

print("\nQuestion 14 Completed: Multi-output Linear Regression model trained successfully with new dataset.")
```
