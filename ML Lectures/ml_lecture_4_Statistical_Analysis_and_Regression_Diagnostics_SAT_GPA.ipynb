```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'../DataSets/RegData.csv')
df.sample(5)
```

```python
col2 = df['GPA']
```

```python
#  plot the distribution plots for the features
import warnings
warnings.filterwarnings('ignore')
def dist2Chrt(c2):
    plt.figure(figsize=(16, 5))
    plt.subplot(1, 2, 2)
    sns.distplot(c2)
    plt.show()
```

```python
def dist2Chart(c2):
    plt.figure(figsize=(16, 5))
    plt.subplot(1, 2, 2)
    sns.histplot(c2)
    plt.show()
```

```python
dist2Chrt(col2)
dist2Chart(col2)
```

```python
sns.boxplot(col2)
```

<!-- Finding the IQR -->

```python
q1 = col2.quantile(0.25)
q2 = col2.quantile(0.50)
q3 = col2.quantile(0.75)
print("Min = 0.0")
print("percentile 25% = ", q1)
print("percentile 50% = ", q2)
print("percentile 75% = ", q3)
print("Max = 100.0")
```

```python
iqr = q3-q1
float(iqr)
```

```python
iqrVal = 1.5 * iqr
lower_limit = q1 - iqrVal
upper_limit = q3 + iqrVal
print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)
```

```python
# Finding the outliners
df[col2 < lower_limit].count()
```

```python
df[col2 > upper_limit].count()
```

```python
df[col2 > upper_limit]
```

```python
# Trimming the Outliners
```

```python
new_df = df[(col2 > lower_limit) & (col2 < upper_limit)]
new_df.shape
```

```python
sns.boxplot(new_df["GPA"])
```

```python
new_df.to_csv('../DataSets/RegData_new.csv')
```
