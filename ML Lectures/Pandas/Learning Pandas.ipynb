# Dataframe Merging

```python
df3 = pd.DataFrame({
    "name":["MAS", "MS"],
    "marks": [1, 2],
})
```

```python
df3
```

```python
df4 = pd.DataFrame({
    "name":["MAS", "MS"],
    "roll": [1233, 2232],
})
```

```python
df4
```

```python
pd.merge(df3, df4, on="name")
```
