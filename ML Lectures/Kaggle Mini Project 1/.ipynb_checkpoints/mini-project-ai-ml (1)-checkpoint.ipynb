```python
import pandas as pd

pd.set_option('display.max_columns', None)
# this below code will ignore any warnings appear after running the cell.
import warnings
warnings.filterwarnings('ignore')
```

# 1. Loading the Data

```python
# We loaded both CSV files into pandas DataFrames.

test_data = pd.read_csv('DataSets/test.csv')
train_data = pd.read_csv('DataSets/train.csv')
```

```python
train_data.head()
```

# 2. Checking & Handling Missing Values (Imputation)

Fill numerical columns with median (LotFrontage, MasVnrArea, GarageYrBlt, etc.)

Fill categorical columns with mode (for very few missing values)

Fill many columns (Alley, FireplaceQu, PoolQC, etc.) with 'None'

```python
# This line of code prints all columns where there are missing values in train data set.
train_cols = train_data.columns
missing_train_cols = []
missing_train_val = []
for col in train_cols:
    if train_data[col].isnull().sum() > 0:
        missing_train_cols.append(col)
        missing_train_val.append(train_data[col].isnull().sum())
        print(f"column {col} have {train_data[col].isnull().sum()} missing values.")
# print(missing_train_cols,missing_train_val)
```

```python
# This line of code prints all columns where there are missing values in test data set
test_cols = test_data.columns
missing_test_cols = []
missing_test_val = []
for col in test_cols:
    if test_data[col].isnull().sum() > 0:
        missing_test_cols.append(col)
        missing_test_val.append(test_data[col].isnull().sum())
        print(f"column {col} have {test_data[col].isnull().sum()} missing values.")
# print(missing_train_cols,missing_train_val)
```

# Imputation Of Missing Values (Most Imp.)

```python
# Here I have filled numerical values with median
train_data['LotFrontage'].fillna(train_data['LotFrontage'].median(), inplace=True)
train_data['MasVnrArea'].fillna(train_data['MasVnrArea'].median(), inplace=True)
train_data['GarageYrBlt'].fillna(train_data['GarageYrBlt'].median(), inplace=True)

# Here I have filled categorical values with mode (there were few missing values)
train_data['Electrical'].fillna(train_data['Electrical'].mode()[0], inplace=True)

# Here I have filled categorical values with median
bsmt_cols = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']
for col in bsmt_cols:
    train_data[col].fillna('None', inplace=True)

# Filled Garage features having missing values with "None" that means "No Garage"
garage_cols = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']
for col in garage_cols:
    train_data[col].fillna('None', inplace=True)

# columns having high missing values replaced with None as well.
high_missing_cols = ['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature']
for col in high_missing_cols:
    train_data[col].fillna('None', inplace=True)

# Special case
train_data['MasVnrType'].fillna('None', inplace=True)

# Check if any missing values remain
print(train_data.isnull().sum().sort_values(ascending=False).head(10))
```

```python
# Here I am filling numerical columns using median from training data
# (we use train data stats so model doesn't learn from test data)
num_cols = [
    'LotFrontage', 'MasVnrArea', 'GarageYrBlt',
    'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF',
    'BsmtFullBath', 'BsmtHalfBath',
    'GarageCars', 'GarageArea'
]

for col in num_cols:
    test_data[col].fillna(train_data[col].median(), inplace=True)
# Here I am filling categorical columns with very few missing values using mode
# again taking mode from training data for consistency
cat_mode_cols = [
    'MSZoning', 'Utilities', 'Exterior1st', 'Exterior2nd',
    'KitchenQual', 'Functional', 'SaleType'
]
for col in cat_mode_cols:
    test_data[col].fillna(train_data[col].mode()[0], inplace=True)
# Here missing basement-related values actually mean the house has no basement
# so instead of guessing, I simply mark them as "None"
bsmt_cols = [
    'BsmtQual', 'BsmtCond', 'BsmtExposure',
    'BsmtFinType1', 'BsmtFinType2'
]
for col in bsmt_cols:
    test_data[col].fillna('None', inplace=True)


# Same idea for garage features → missing means no garage exists
garage_cols = [
    'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'
]

for col in garage_cols:
    test_data[col].fillna('None', inplace=True)


# These columns have a lot of missing values and mostly represent optional features
# so I replaced missing values with "None" to indicate absence
high_missing_cols = [
    'Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'
]

for col in high_missing_cols:
    test_data[col].fillna('None', inplace=True)


# Special case: missing MasVnrType usually means no masonry veneer
test_data['MasVnrType'].fillna('None', inplace=True)


# Final check to make sure no major missing values are left
print(test_data.isnull().sum().sort_values(ascending=False).head(10))
```

# sklearn for Linear Regression, train-test split, and R² score

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score, mean_squared_error
```

# 1) Apply One Hot Encoding on categorical columns

What happened: All text columns (MSZoning, Neighborhood, GarageType, etc.) were converted into binary (0/1) columns.
    
Why we did it: Linear models can only understand numbers, not text.
    
Why this step was required: Without converting text to numbers, the model cannot be trained.

```python
# Identify categorical columns
categorical_cols = train_data.select_dtypes(include=['object']).columns.tolist()

# Separate features and target
X = train_data.drop(['SalePrice', 'Id'], axis=1)
y = train_data['SalePrice']

# One-hot encoding
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# Apply same encoding on test data
test_encoded = pd.get_dummies(test_data.drop('Id', axis=1), 
                              columns=categorical_cols, 
                              drop_first=True)

# Align columns (very important)
test_encoded = test_encoded.reindex(columns=X_encoded.columns, fill_value=0)

print("Step 1 Completed: One-Hot Encoding Done")
print(f"Train shape after encoding: {X_encoded.shape}")
print(f"Test shape after encoding : {test_encoded.shape}\n")
```

# 2) Train-Test Split (for model evaluation)

What happened: Training data was split into 80% training + 20% validation.
    
Why we did it: To check model performance on data it has never seen before.
    
Why required: This gives us an honest R² score before submitting to Kaggle. Without it, we would have no idea if the model is good.

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

print("Step 2 Completed: Train-Test Split Done")
print(f"Training samples: {X_train.shape[0]} | Validation samples: {X_val.shape[0]}\n")
```

# 3) Train Linear Regression model using train dataset → We use Ridge (improved version of Linear Regression) + Scaling

What happened:

Target (SalePrice) was log-transformed.
Features were scaled using StandardScaler.
We trained Ridge Regression (improved version of Linear Regression) instead of plain Linear Regression.

Why we did it:

House prices are skewed → log transformation makes the distribution more normal.
Features have very different scales → StandardScaler brings them to same scale.
Ridge adds regularization → prevents overfitting and gives better generalization.

Why required: Plain Linear Regression gave a bad score (2.01). These improvements make the model much stronger.

```python
# Log transform target (critical for reducing score)
y_train_log = np.log1p(y_train)

# Improved model: Ridge + StandardScaler (much better than plain LinearRegression)
model = make_pipeline(
    StandardScaler(),
    Ridge(alpha=15)
)

model.fit(X_train, y_train_log)

print("Step 3 Completed: Model Trained (Ridge with Scaling)\n")
```

# 4) Evaluate the model performance using R² as evaluation metric

What happened: You calculated R² score on validation data → 0.8727

Why we did it: Your original notebook instructions specifically asked to use R² as the evaluation metric.
    
Why required: To measure how well the model explains house price variation.

```python
# Predict on validation set (in log scale)
val_pred_log = model.predict(X_val)
y_val_log = np.log1p(y_val)

# R² Score (as Sir asked)
r2 = r2_score(y_val_log, val_pred_log)
print(f"Step 4: R² Score on Validation Data = {r2:.4f}")

# Also show RMSE for better understanding (lower is better)
rmse = np.sqrt(mean_squared_error(y_val_log, val_pred_log))
print(f"Validation RMSE (log scale) = {rmse:.5f}\n")
```

# 5) Generate predictions using the test_data

What happened: Model made predictions on the real test data and converted them back from log scale.

Why required: Kaggle scores only the predictions on the test set.

```python
test_pred_log = model.predict(test_encoded)

# Convert back from log scale to original SalePrice
test_predictions = np.expm1(test_pred_log)

# Safety: No negative prices
test_predictions = np.maximum(0, test_predictions)

print("Step 5 Completed: Predictions generated on test data\n")
```

# 6) Save the generated predictions in a csv file

```python
submission = pd.DataFrame({
    'Id': test_data['Id'],
    'SalePrice': test_predictions
})

submission.to_csv('submission_improved.csv', index=False)

print("Step 6 Completed: Submission file saved as 'submission_improved.csv'")
print(submission.head())

print("\n✅ Now download 'submission_improved.csv' and submit it on Kaggle!")
print("After submission, note your new score and ranking.")
```
