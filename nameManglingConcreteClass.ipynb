```python
class Vehicle:
# This is a special method called a Constructor.
# It runs when you create a new object.
    def __init__(self, wheels, make, model, color):
        self.__vehType = "Vehicle" # __ Private instance attribute
        self.__wheels = wheels # __ Private instance attribute
        self._make = make # _ Protected instance attribute
        self._model = model # _ Protected instance attribute
        self.color = color # Public instance attribute
    def set_vehType(self, vehType):
        self.__vehType = vehType # A public method to set the private attribute
    def get_vehType(self):
        return self.__vehType # A public method to access the private attribut
    def get_wheels(self):
        return self.__wheels # A public method to access the private attribute
    def get_info(self):
        infor = f"{self.color} {self.__wheels}-wheeler, {self._make} {self._model}"
        return infor

    def move(self, txt="The"):
        print(f"{txt} {self.get_vehType()} is moving.")
```

```python
# An abstract class is basically:

# A blueprint that you are NOT allowed to use directly.

# It exists only to be inherited (extended) by other classes.

# Why Abstract Classes Exist (Real Reason)
# Without abstract classes:

# You might forget to implement move()

# Your design becomes inconsistent

# With abstract classes:

# 👉 Python forces discipline

# Mental Model (Remember This)

# Think of Vehicle like:

# “Every vehicle must move, but HOW it moves depends on the type.”

# Car → driven
# Bike → ridden

# So:

# 👉 Abstract class = “rule maker”
# 👉 Child classes = “rule implementers”
```

```python
# ABC = Abstract Base Class
# This makes Vehicle special

from abc import ABC, abstractmethod
class Vehicle(ABC):
# This is a special method called a Constructor.
# It runs when you create a new object.
    def __init__(self, wheels, make, model, color):
        self.__vehType = "Vehicle" # __ Private instance attribute
        self.__wheels = wheels # __ Private instance attribute
        self._make = make # _ Protected instance attribute
        self._model = model # _ Protected instance attribute
        self.color = color # Public instance attribute
    def set_vehType(self, vehType):
        self.__vehType = vehType # A public method to set the private attribute
    def get_vehType(self):
        return self.__vehType # A public method to access the private attribut
    def get_wheels(self):
        return self.__wheels # A public method to access the private attribute
    def get_info(self):
        infor = f"{self.color} {self.__wheels}-wheeler, {self._make} {self._model}"
        return infor
        
    @abstractmethod
    def move(self, txt="The"):
        pass


# It means:

# "Any class that inherits me MUST define move() — no excuses."
```

```python
veh = Vehicle(4, "Toyota", "Corolla", "Blue")
```

```python
class Car(Vehicle):
    def __init__(self, wheels, make, model, color):
        super().__init__(wheels, make, model, color)
        self.set_vehType("Car")

    def move(self, txt="The"):
        print(f"{txt} {self.get_vehType()} is being driven.")

# Child classes implement the missing behavior
# Same method name (move)
# 👉 Different behavior

# This is polymorphism
```

```python
class MotorBike(Vehicle):
    def __init__(self, wheels, make, model, color):
        super().__init__(wheels, make, model, color)
        self.set_vehType("Bike")
    def move(self, txt="The"):
        print(f"{txt} {self.get_vehType()} is being ridden.")

# Child classes implement the missing behavior
# Same method name (move)
# 👉 Different behavior

# This is polymorphism
```
