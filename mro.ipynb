```python
class A:
    def process(self):
        print('A process()')
```

```python
class B:
    def process(self):
        print('B process()')
```

```python
# depends on which process is given first in the attribute
class C(B, A): 
    def process(self):
        print('C process()')
```

```python
obj = C()
obj.process()
print(C.mro())
```

```python
class A:
    def process(self):
        print("A process()")

class B(A):
    pass

class C(A):
    def process(self):
        print("C process()")

class D(B, C):
    pass
```

```python
obj = D()
obj.process()
print(D.mro())
```

```python

# c cant access the super class of B directly
class A:
    def process(self):
        print('A process()')

class B(A):
    def process(self):
        print('B process()')

class C(A, B):
    pass

```

```python
obj = C()
obj.process()
```
