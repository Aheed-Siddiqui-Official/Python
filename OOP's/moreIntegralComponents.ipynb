```python
class Engine:
    def __init__(self, engnID):
        self.__engnID = engnID
        self.__estatus = "Working"

    def set_estatus(self, estatus):
        self.__estatus = estatus

    def get_estatus(self):
        return self.__estatus

    def get_engnID(self):
        return self.__engnID
```

```python
class Battery:
    def __init__(self, btryID):
        self.__btryID = btryID
        self.__btrystatus = "Working"

    def set_btrystatus(self, btrystatus):
        self.__btrystatus = btrystatus

    def get_btryID(self):
        return self.__btryID

    def get_btrystatus(self):
        return self.__btrystatus
```

```python
class AC:
    def __init__(self, acID):
        self.__acID = acID
        self.__acstatus = "Off"         

    def set_acstatus(self, acstatus):
        self.__acstatus = acstatus

    def get_acID(self):
        return self.__acID

    def get_acstatus(self):
        return self.__acstatus
```

```python
class Vehicle:
    def __init__(self, vehID, wheels, make, model, color, engine, battery, ac=None):
        self._vehType = "Vehicle"
        self.__vehID = vehID
        self.__wheels = wheels
        self.__make = make
        self.__model = model
        self.__color = color
        self.__engine = engine
        self.__battery = battery      
        self.__ac = ac   
        self.__vStatus = "Working"        

    def set_vehType(self, vehType):
        self._vehType = vehType

    def set_color(self, color):
        self.__color = color

    def set_engn_status(self, sts="Working"):
        self.__engine.set_estatus(sts)

    def set_btry_status(self, sts="Working"):
        self.__battery.set_btrystatus(sts)

    def set_ac_status(self, sts="Working"):
        if self.__ac is not None:
            self.__ac.set_acstatus(sts)

    def get_vehType(self):
        return self._vehType

    def get_engn(self):
        return self.__engine

    def get_engnID(self):
        return self.__engine.get_engnID()

    def get_engn_status(self):
        return self.__engine.get_estatus()

    def get_btry(self):
        return self.__battery
    
    def get_btryID(self):
        return self.__battery.get_btryID()

    def get_btry_status(self):
        return self.__battery.get_btrystatus()

    def get_ac(self):
        return self.__ac
    
    def get_acID(self):
        if self.__ac is None:
            return None
        return self.__ac.get_acID()

    def get_ac_status(self):
        if self.__ac is None:
            return "No AC installed"
        return self.__ac.get_acstatus()
    
    def get_vStatus(self):
        return self.__vStatus

    def get_info(self):
        return f"{self.__color} {self.__wheels}-wheeler, {self.__make} {self.__model} having " \
               f"Engine:{self.get_engnID()}, Battery:{self.get_btryID()}, AC:{self.get_acID()}"

    def move(self, txt="The"):
        if self.get_engn_status() != "Working" or \
           self.get_btry_status() != "Working" or \
           (self.__ac is not None and self.get_ac_status() != "Working"):
            self.__vStatus = "Problem"
            raise Exception("The Car couldn't start: Vehicle Problem")
        
        self.__vStatus = "Working"
        return f"{txt} vehicle is moving."
```

```python
class Car(Vehicle):
    def __init__(self, vehID, wheels, make, model, color, engine, battery, ac=None):
        super().__init__(vehID, wheels, make, model, color, engine, battery, ac)
        self.set_vehType("Car")

    def move(self, txt="The"):
        if self.get_engn_status() != "Working" or \
           self.get_btry_status() != "Working" or \
           (self.get_ac() is not None and self.get_ac_status() != "Working"):
            self.__vStatus = "Problem"
            raise Exception("The Car couldn't start: Vehicle Problem")
        
        self.__vStatus = "Working"
        return f"{txt} Car is being driven."
```

```python
engn3 = Engine(3)
btry3 = Battery(3)
ac3 = AC(3)

my_car = Car(1, 4, "Toyota", "Corolla", "Blue", engn3, btry3, ac3)
```

```python
info = my_car.get_info()
print(info)
```
