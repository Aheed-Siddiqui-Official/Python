```python
class Counter:
    # static works like a global variable and it remain i the memory for the time until the class objects exits
    _instances = 0

    def __init__(self, name):
        Counter._instances += 1
        self.name = name
        print(f"Created {self.name}, total instances:{Counter._instances}")

    def __del__(self):
        Counter._instances -= 1
        print(f"Created {self.name}, total instances:{Counter._instances}")
```
