```python
class Product:
    def __init__(self, name, price, description):
        self.name = name
        self.price = price
        self.description = description


class PhysicalProduct(Product):
    def __init__(self, name, price, description, weight, shipping_cost):
        super().__init__(name, price, description)
        self.weight = weight
        self.shipping_cost = shipping_cost


class DigitalProduct(Product):
    def __init__(self, name, price, description, file_size, download_link):
        super().__init__(name, price, description)
        self.file_size = file_size
        self.download_link = download_link


class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_product(self, product):
        self.items.append(product)

    def calculate_total(self):
        total = 0
        for item in self.items:
            total += item.price

            if isinstance(item, PhysicalProduct):
                total += item.shipping_cost

        return total
```

```python
p1 = PhysicalProduct("Phone", 500, "Smartphone", 0.5, 20)
p2 = DigitalProduct("Course", 50, "Python Course", "2GB", "link")

cart = ShoppingCart()
cart.add_product(p1)
cart.add_product(p2)

print(cart.calculate_total())
```
