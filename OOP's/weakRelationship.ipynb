```python
class Engine:
    def __init__(self, engnID):
        self.__engnID = engnID
        self.__estatus = "Working"

    def set_estatus(self, estatus):
        self.__estatus = estatus

    def get_estatus(self):
        return self.__estatus

    def get_engnID(self):
        return self.__engnID
```

```python
class AC:
    def __init__(self, acID, status="Off"):
        self.__acID = acID
        self.__status = status

    def set_status(self, status):
        self.__status = status

    def get_status(self):
        return self.__status

    def get_acID(self):
        return self.__acID
```

```python
class Vehicle:
    def __init__(self, vehID, wheels, make, model, color, engine ,ac = None):
        self._vehType = "Vehicle"
        self.__vehID = vehID
        self.__wheels = wheels
        self.__make = make
        self.__model = model
        self.__color = color
        self.__engine = engine
        self.__ac = ac # weak relationship

    def set_vehType(self, vehType):
        self._vehType = vehType

    def set_ac(self, ac):
        self.__ac = ac

    def get_ac(self):
        return self.__ac

    def set_color(self, color):
        self.__color = color

    def set_engn_status(self, status="Working"):
        self.__engine.set_estatus(status)

    def get_vehType(self):
        return self._vehType

    def get_engn(self):
        return self.__engine

    def get_engnID(self):
        return self.__engine.get_engnID()

    def get_engn_status(self):
        return self.__engine.get_estatus()

    def get_info(self):
        return f"{self.__color} {self.__wheels}-wheeler, {self.__make} {self.__model} having Engine:{self.get_engnID()}"

    def move(self, txt="The"):
        if self.get_engn_status() != "Working":
            raise Exception("The Car couldn't start: Engine Problem")
        return f"{txt} vehicle is moving."
    
    def get_ac_status(self):
        if self.__ac is None:
            return "No AC installed"
        return self.__ac.get_status()
```

```python
class Car(Vehicle):
    def __init__(self, vehID, wheels, make, model, color, engine):
        super().__init__(vehID, wheels, make, model, color, engine)
        self.set_vehType("Car")

    def move(self, txt="The"):
        if self.get_engn_status() != "Working":
            raise Exception("The Car couldn't start: Engine Problem")
        return f"{txt} Car is being driven."
```

```python
engn1 = Engine(1)
ac1 = AC(101, "Off")   # independent object

my_car = Car(1, 4, "Toyota", "Corolla", "Blue", engn1)

my_car.set_ac(ac1)
```

```python
print("AC Status:", my_car.get_ac_status())

ac1.set_status("On")

print("AC Status after turning on:", my_car.get_ac_status())
```
