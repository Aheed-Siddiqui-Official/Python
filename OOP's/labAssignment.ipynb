```python
class ElectricDevice:
    def __init__(self, brand, model, battery, power, weight):
        self._brand = brand
        self._model = model
        self._battery = battery
        self._power = power
        self._weight = weight

    def charge(self, amount):
        self._battery += amount
        if self._battery > 100:
            self._battery = 100
        print(self._brand, "charged to", self._battery, "%")

    def turn_on(self):
        print("Device is turning on")

    def turn_off(self):
        print("Device is turning off")


class Mobile(ElectricDevice):
    def turn_on(self):
        print(self._brand, "Mobile is ON")

    def turn_off(self):
        print(self._brand, "Mobile is OFF")

class Tablet(ElectricDevice):
    def turn_on(self):
        print(self._brand, "Tablet is ON")

    def turn_off(self):
        print(self._brand, "Tablet is OFF")

class PC(ElectricDevice):
    def turn_on(self):
        print(self._brand, "PC is starting")

    def turn_off(self):
        print(self._brand, "PC is shutting down")

class SmartWatch(ElectricDevice):
    def turn_on(self):
        print(self._brand, "SmartWatch is ON")

    def turn_off(self):
        print(self._brand, "SmartWatch is OFF")

devices = [
    Mobile("Samsung", "S23", 50, False, 0.2),
    Tablet("Apple", "iPad", 70, False, 0.5),
    PC("Dell", "XPS", 100, False, 2.0),
    SmartWatch("Huawei", "GT", 30, False, 0.1)
]

for d in devices:
    d.turn_on()
    d.charge(10)
    d.turn_off()
    print("-----")
```
