```python
class Vehicle:
    # This is an attribute (Class Variable) common to all cars
    __vehType = "Vehicle"
    # This is a special method called a Constructor.
    # It runs when you create a new Car object.
    def __init__(self, wheels, make, model, color):
        self.__wheels = wheels    # Instance attribute
        self.__make = make        # Instance attribute
        self.__model = model      # Instance attribute
        self.color = color   

    # Functions
    def set_vehType(self, vehType):
        self.__vehType = vehType
        
    def get_vehType(self):
        return self.__vehType

    def get_wheels(self):
        return self.__wheels

    def get_make(self):
        return self.__make

    def get_model(self):
        return self.__model

    def get_info(self):
        info = f"Wheels: {self.__wheels}, Make: {self.__make}, Model: {self.__model}, Color: {self.color}"
        return info
    # Procedure
    def move(self, txt="The"):
        print(f"{txt} {self.get_vehType()} is moving.")
```

```python
class Car(Vehicle):
    def __init__(self, wheels, make, model, color):
            super().__init__(wheels, make, model, color)
            self.set_vehType("Car")

class Motorcycle(Vehicle):
    def __init__(self, wheels, make, model, color):
            super().__init__(wheels, make, model, color)
            self.set_vehType("MotorBike")
        
```
